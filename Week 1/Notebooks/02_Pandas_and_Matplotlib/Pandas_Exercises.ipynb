# Pandas Exercises

```python
import pandas as pd

url = "https://raw.githubusercontent.com/BeaverWorksMedlytics2020/Data_Public/master/NotebookExampleData/Week1/diabetes.csv"
names = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']
```

Read in csv above to a pandas dataframe

```python
# YOUR CODE HERE
# YOUR CODE HERE
data = pd.read_csv(url, names=names)
```

Print out the age and class of the 100th patient

```python
# YOUR CODE HERE
print(f"Age of 100th patient: {data.loc[99, 'age']}")
print(f"Class of 100th patient: {data.loc[99, 'class']}")
```

Write code to figure out how many people are classified as not having diabetes, class is 0

```python
# YOUR CODE HERE
non_diabetic = data[data['class'] == 0].shape[0]
print(f"Number of people without diabetes (class 0): {non_diabetic}")
```

Write code to figure out how many people are classified as having diabetes, class is 1

```python
# YOUR CODE HERE
diabetic_count = data[data['class'] == 1].shape[0]
print(f"Number of people with diabetes (class 1): {diabetic_count}")
```

What percent of people have diabetes in the study

```python
# YOUR CODE HERE
percent_diabetic = (diabetic_count / data.shape[0]) * 100
print(f"Percent of people with diabetes: {percent_diabetic}%")
```

```python
data.corr()
```

What column has the highest correlation with class? Find the mean() value of that column

```python
# YOUR CODE HERE
correlation_with_class = data.corr()['class'].abs().sort_values(ascending=False)

highest_corr_column = correlation_with_class.index[1]

mean_of_column = data[highest_corr_column].mean()

print(f"Column with the highest correlation with 'class': {highest_corr_column}")
print(f"Mean of '{highest_corr_column}': {mean_of_column:.2f}")
```

Create a new dataframe with only people that are above the mean found in the above cell.

```python
# YOUR CODE HERE
df_above_plas_mean = data[data[highest_corr_column] > mean_of_column]
print(df_above_plas_mean.head())
```

Create a new dataframe that only includes people who are above the average value for each column in the dataset. In other words, find the average value for each column, and then find which users have values above all of these average column values.

```python
column_means = data.mean()# YOUR CODE HERE
df_above_all_means = data[data > column_means].dropna()
print(df_above_all_means.head())
```
