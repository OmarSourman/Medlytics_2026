```python
above_mean_data= data[(data['preg'] > data['preg'].mean())
& (data['plas'] > data['plas'].mean()) &
(data['pres'] > data['pres'].mean()) &
(data['skin'] > data['skin'].mean()) &
(data['test'] > data['test'].mean() )& (data['mass'] > data['mass'].mean()) & (data['pedi'] > data['pedi'].mean()) & (data['age'] > data['age'].mean()) &( data['class'] > data['class'].mean())]
```
